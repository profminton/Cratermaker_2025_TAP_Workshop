```python
import pandas as pd
from cratermaker import Crater, Simulation

skipnum = 500
craters = pd.read_csv("ohman3.csv")

sim = Simulation(simdir="real_craters",gridlevel=7, dorays=False, raster_format="vtp", vector_format="vtp")
sim.save()

craterlist = []
for index, row in craters.iterrows():
    craterlist.append(
        Crater.maker(final_diameter=row["Diameter"], location=(row["Lon"], row["Lat"]))
    )
    # print(row["Name"])
    if index % skipnum == 0 or index == len(craters) - 1:
        sim.emplace(craterlist)
        sim.interval_number += 1
        sim.save()
        craterlist = []


```
